```python
def check_prod(prod_id, quant):     #function checks if the Prod id and quantity are appropriate
    if prod_id not in curr_inventory:
        print("Entered product ID is incorrect. Please try again")
        return False
    
    if curr_inventory[prod_id]['quant'] == 0:
        print("We're sorry. This product is out of stock")
        return False
    
    if curr_inventory[prod_id]['quant'] < quant:
        print("The number of items needed exceeds the current inventory stock. Please enter a smaller quantity")
        return False
        
    return True



import json
from datetime import datetime as dt

f = open('records.json', 'r')   #open json file to show menu of current inventory items
txt = f.read()
f.close()

curr_inventory  = json.loads(txt)

print("**********************************************************WELCOME**********************************************************")

print("YOUR MENU IS AS FOLLOWS".center(120))


print("\n{:<13} {:<25} {:<10} {:<15} {:<13} {:<10} {:<10}".format('PRODUCT-ID', 'PRODUCT-NAME', 'PRICE', 'MFG', 'EXP', 'CATEGORY', 'QUANTITY IN STOCK'))

for product in curr_inventory:
    print("{:<13} {:<25} {:<10} {:<15} {:<14} {:<15} {:<10}".format(product,curr_inventory[product]['prod_name'],curr_inventory[product]['MRP'],curr_inventory[product]['mfg'],curr_inventory[product]['exp'],curr_inventory[product]['category'],curr_inventory[product]['quant']))
    


u_inp = input("\n\nPress 'f' if you want to filter products by category or press any other key to continue: ")

if(u_inp == 'f'):  #filter products by category
    flag = 0
    while(flag == 0):  #prompting user to enter category to filter products until category is valid
        
        filt = input("Enter the category you're interested in: ").lower()
        
        for product in curr_inventory:
            if curr_inventory[product]['category'] == filt:
                print("\n{:<13} {:<25} {:<10} {:<15} {:<14} {:<15} {:<10}".format(product,curr_inventory[product]['prod_name'],curr_inventory[product]['MRP'],curr_inventory[product]['mfg'],curr_inventory[product]['exp'],curr_inventory[product]['category'],curr_inventory[product]['quant']))
                flag=1

        if flag==0:
            print("\nYou did not enter a valid category. Please try again\n")

user_prods = {}  #shopping cart
while(1):
    inp_id = input("Enter Product ID of the product you want to purchase: ")
    quant =  int(input("Enter quantity required by you: "))
    if quant <=0:
        print("Please enter a valid quantity")
        continue
        
    if(check_prod(inp_id,quant)):   #function to check if the prod id and quantity are appropriate
        user_prods[inp_id] = quant
        choice = input("Would you like to add more items to your cart? Press 'y' if yes or press any other key to proceed to checkout: ")
        if choice == 'y':
            continue
    
        else:
            break
            
            
            
            
            
#-------------------------------------------------------updating inventory-------------------------------------------------------
for product in user_prods:
    curr_inventory[product]['quant']-= user_prods[product]
    
    
f = open("records.json",'w')
js = json.dumps(curr_inventory)
f.write(js)
f.close()



#------------------------------------------------------updating sales file------------------------------------------------------


f=open("sales.json",'r')
txt = f.read()
f.close()
sales = json.loads(txt)
date = str(dt.now().strftime("%Y-%m-%d"))

arr=[]
for p in user_prods:
    name = curr_inventory[p]['prod_name']
    ID = p
    quant = user_prods[p]
    amount = quant * curr_inventory[p]['MRP']
    
    arr.append({'name':name, 'id':ID, 'amount':amount, 'quant':quant})
    

if date not in sales:
    sales[date] = arr
    
else:
    for a in arr:
        sales[date].append(a)
        


f = open("sales.json",'w')
txt = json.dumps(sales)
f.write(txt)
f.close()



#------------------------------------------------------------------------------------------------------------------------------




print("\n\nYour shoppping cart has the following: ")
print("\n{:<25} {:<15} {:<10} {:<10}".format('Product','Quantity', 'Price', 'Sub-Total'))
total=0
for prod in user_prods:
    print("{:<25} {:<15} {:<10} {:<10}".format(curr_inventory[prod]['prod_name'],user_prods[prod],curr_inventory[prod]['MRP'],curr_inventory[prod]['MRP']*user_prods[prod]))
    total+=curr_inventory[prod]['MRP']*user_prods[prod]

    
print("\n\n\n")
print("PROCEEDING TO CHECKOUT...".center(120))
print("\n\n\n")
print("---------------------------------------------------------BILL---------------------------------------------------------")
print("\n{:<25} {:<25} {:<15} {:<10} {:<10}".format(' ','Product','Quantity', 'Price', 'Sub-Total'))
for prod in user_prods:
    print("{:<25} {:<25} {:<15} {:<10} {:<10}".format(' ',curr_inventory[prod]['prod_name'],user_prods[prod],curr_inventory[prod]['MRP'],curr_inventory[prod]['MRP']*user_prods[prod]))
    

print("\n\n{:<25} {:<2} {:<13} {:<2}".format(' ',"TOTAL AMOUNT TO BE PAID: Rs.",total,str(dt.now().strftime("%Y-%m-%d, %H:%M"))))
print("\n----------------------------------------------------------------------------------------------------------------------")




```
